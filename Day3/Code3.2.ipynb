```python
import requests
data = requests.get("https://raw.githubusercontent.com/00Dabide/Advent_of_Code_2023/main/Day3/Text3.1.txt").text.split("\n")
data
```

```python
EveryStar = {}

# for every row
for row in range(len(data)):
    rowNum = []
    place = 0

    # for every character
    while place <= len(data[row])-1:
        area = []
        # if character is *
        if data[row][place] == "*":
            if place-1 != -1:
                area.append(place-1)
            area.append(place)
            if place+1 != len(data[row])-1:
                area.append(place+1)
            rowNum.append(area)
        place += 1
    EveryStar[row] = rowNum
EveryStar
```

```python
EveryNum = {}
for row in range(len(data)):
    rowNum = []
    place = 0
    while place <= len(data[row])-1:
        NumChar = ""
        Number = 0
        area = []
        if data[row][place].isdigit():
            i = 0
            while data[row][place+i].isdigit():
                NumChar += data[row][place+i]
                area.append(place+i)
                i += 1
                if place+i > len(data[row])-1:
                    break
            Number = int(NumChar)
            rowNum.append([Number,area])
            place += i-1
        place += 1
    EveryNum[row] = rowNum
EveryNum
```

```python
# Between
Mult = 0

for row in range(1, len(data)-1):
    for Symbol in EveryStar[row]:
        Count = 0
        FoundNum = []
        for area in Symbol:
            
            # Above it
            for NUM in EveryNum[row-1]:
                if area in NUM[1]:
                    if NUM[0] not in FoundNum:
                        Count += 1
                        FoundNum.append(NUM[0])

            # Same row
            for NUM in EveryNum[row]:
                if area in NUM[1]:
                    if NUM[0] not in FoundNum:
                        Count += 1
                        FoundNum.append(NUM[0])

            # Below it
            for NUM in EveryNum[row+1]:
                if area in NUM[1]:
                    if NUM[0] not in FoundNum:
                        Count += 1
                        FoundNum.append(NUM[0])
        if Count == 2:
            Mult += (FoundNum[0] * FoundNum[1])
```
