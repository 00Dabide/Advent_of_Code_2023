```python
import requests
data = requests.get("https://raw.githubusercontent.com/00Dabide/Advent_of_Code_2023/main/Day4/Text4.1.txt").text.split("\n")
data
```

```python
data[0].split(":")[1].split("|")[0].split(" ")
```

```python
WinNUM = []
MyNUM = []

for card in range(len(data)):
    WinNUM.append([x for x in data[card].split(":")[1].split("|")[0].split(" ") if x != ""])
    MyNUM.append([x for x in data[card].split(":")[1].split("|")[1].split(" ") if x != ""])
```

```python
MyNUM
```

```python
Sum = 0
for card in range(len(data)):
    n = 0
    for number in MyNUM[card]:
        if number in WinNUM[card]:
            n += 1
    if n > 0:
        Sum += 2**(n-1)
Sum
```
