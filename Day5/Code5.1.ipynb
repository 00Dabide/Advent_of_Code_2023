```python
import requests
data = requests.get("https://raw.githubusercontent.com/00Dabide/Advent_of_Code_2023/main/Day5/Text5.1.txt").text.split("\n")
data
```

```python
seeds = data[0].split(" ")
seeds.remove("seeds:")
seeds = [int(x) for x in seeds]
seeds
```

```python
maps = data[2::]
maps
```

```python
maps2 = []
dummy = []
for i in maps:
    if i == "":
        maps2.append(dummy)
        dummy = []
    else:
        dummy.append(i)
maps2.append(dummy)

for i in maps2:
    i.pop(0)
```

```python
maps2
```

```python
maps2[0][0].split(" ")
```

```python
# 7 maps = 7 deepness

locations = []
for seed in seeds:
    treasure = seed
    for map in maps2:
        for row in map:
            row = row.split(" ")
            if treasure in range(int(row[1]),int(row[1])+int(row[2])):
                treasure += int(row[0]) - int(row[1])
                break

    locations.append(treasure)
```

```python
min(locations)
```
